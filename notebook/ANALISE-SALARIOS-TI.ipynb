## Analise da consistencia do **df_etl_salary_completo** para análise

```python
print("Analise exploratórias dos dados do df_etl_salary_completo")
df = pd.read_csv('/Workspace/Users/jhenivon@hotmail.com/Projetos/dados/df_etl_salary_completo.csv')
df.info()
df.head(5)
```

## 1️⃣ Qual o nível de experiência mais comum na base?

```python
import plotly.express as px
import pandas as pd

df_count = df['senioridade'].value_counts().reset_index()
df_count.columns = ['senioridade', 'quantidade']

fig = px.bar(
    df_count,
    x='senioridade',
    y='quantidade',
    text='quantidade',
    title='Distribuição de Níveis de Experiência',
    color='senioridade',
    color_discrete_sequence=px.colors.qualitative.Dark2
)

fig.update_traces(texttemplate='%{text}', textposition='outside')
fig.update_layout(
    height=900, 
    template='plotly_dark',
    yaxis_title='Quantidade de Profissionais',
    xaxis_title='Nível de Experiência'
)
fig.show()
```

## 2️⃣ Top 10 países com maior média salarial

```python
import plotly.express as px

# Agrupa e calcula a média salarial por país
df_media = (
    df.groupby('residencia_profissional')['salario_bruto']
      .mean()
      .reset_index()
      .sort_values('salario_bruto', ascending=False)
)

# Gráfico de barras com os 10 países de maior salário médio
fig = px.bar(
    df_media.head(10),
    x='residencia_profissional',
    y='salario_bruto',
    title='Top 10 Países com Maior Salário Médio',
    color='residencia_profissional',
    text='salario_bruto'
)

# Ajustes visuais
fig.update_traces(
    texttemplate='%{text:.2s}',
    textposition='outside',
    textfont=dict(color='white', size=14)
)

fig.update_layout(
    template='plotly_dark',
    xaxis_title='País',
    yaxis_title='Salário Bruto',
    height=700, 
    width=1000, 
    margin=dict(t=80, b=40, l=80, r=80),
    yaxis=dict(range=[0, df_media['salario_bruto'].max() * 1.15])  
)

fig.show()
```

## 4️⃣ Qual o cargo mais bem pago por senioridade?

```python

import plotly.express as px

df_cargo = (
    df.groupby(['cargo','senioridade'])['salario_bruto']
      .mean()
      .reset_index()
)

top_cargos = (
    df_cargo.groupby('cargo')['salario_bruto']
      .mean()
      .sort_values(ascending=False)
      .head(15)
      .index
)

df_top15 = (
    df_cargo[df_cargo['cargo'].isin(top_cargos)]
      .sort_values('salario_bruto', ascending=False)
)

fig = px.bar(
    df_top15,
    y='cargo',
    x='salario_bruto',
    color='senioridade',
    text='salario_bruto', 
    orientation='h',
    title='Top 15 Cargos com Maior Salário Médio por Senioridade (Horizontal)',
)

# Ajustes visuais
fig.update_traces(
    texttemplate='%{text:.2s}',         
    textposition='outside',             
    textfont=dict(color='white', size=14),   
    insidetextfont=dict(color='white', size=14)
)

fig.update_layout(
    template='plotly_dark',
    height=800,   
    width=1200,  
    margin=dict(t=80, b=40, l=80, r=80),
    yaxis={'categoryorder':'total ascending'},
    xaxis_title='Salário Médio',
    yaxis_title='Cargo'
)

fig.show()
```

## 5️⃣ Como varia o salário médio por tamanho da empresa?

```python
import plotly.express as px

# Agrupa e calcula a média salarial por tamanho da empresa
df_empresa = (
    df.groupby('tamanho_empresa')['salario_bruto']
      .mean()
      .reset_index()
      .sort_values('salario_bruto', ascending=True)
)

# Gráfico de barras
fig = px.bar(
    df_empresa,
    x='tamanho_empresa',
    y='salario_bruto',
    color='tamanho_empresa',
    text='salario_bruto',   
    title='Salário Médio por Tamanho da Empresa (Ordenado)',
)

# Ajustes visuais dos textos
fig.update_traces(
    texttemplate='%{text:.2s}',          
    textposition='outside',              
    textfont=dict(color='white', size=14) 
)

# Layout geral
fig.update_layout(
    template='plotly_dark',
    xaxis={'categoryorder':'total descending'},  
    yaxis_title='Salário Médio',
    xaxis_title='Tamanho da Empresa',
    height=700,   
    width=1000,  
    margin=dict(t=80, b=40, l=80, r=80) 
)

fig.show()
```

## 6️⃣ Qual o crescimento salarial ao longo dos anos?


```python
import plotly.express as px

# Média salarial por ano
df_ano = df.groupby('ano')['salario_bruto'].mean().reset_index()

# Gráfico de linha com rótulos
fig = px.line(
    df_ano,
    x='ano',
    y='salario_bruto',
    text='salario_bruto',
    title='Evolução do Salário Médio ao Longo dos Anos',
    markers=True  
)

# Ajustes visuais
fig.update_traces(texttemplate='%{text:.2s}', textposition='top center')
fig.update_layout(
    template='plotly_dark',
    yaxis_title='Salário Médio',
    xaxis_title='Ano'
)

fig.show()
```

## 7️⃣ Há diferença entre salário bruto e salary_usd por região?

Eixo Y escala logarítmica, para mais detalhes pesquisa por "escala logarítmica"

```python
import plotly.express as px
import pandas as pd


df_diff = (
    df.groupby('residencia_profissional')[['salario_bruto','salary_usd']]
      .mean()
      .reset_index()
      .sort_values('salario_bruto', ascending=False)
)

df_diff = df_diff.head(15)

df_melt = df_diff.melt(
    id_vars='residencia_profissional',
    value_vars=['salario_bruto','salary_usd'],
    var_name='Tipo_Salario',
    value_name='Valor_Medio'
)

fig = px.bar(
    df_melt,
    x='residencia_profissional',
    y='Valor_Medio',
    color='Tipo_Salario',
    title='Comparativo Consolidado: Salário Bruto vs USD por País (Top 15)',
    color_discrete_map={
        'salario_bruto':'rgba(99,110,250,0.7)',
        'salary_usd':'rgba(239,85,59,0.7)'
    },
    barmode='group'
)

fig.update_layout(
    template='plotly_dark',
    xaxis={'categoryorder':'total descending'},
    xaxis_title='País',
    yaxis_title='Valor Médio',
    legend_title_text='Tipo de Salário',
    height=700   
)

fig.update_yaxes(type='log')

fig.update_traces(texttemplate='%{y:.2s}', textposition='outside')

fig.show()
```
